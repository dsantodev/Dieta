## 🥗 Guida agli Ingredienti

La chetogenica si basa su un alto contenuto di grassi, un moderato apporto proteico e pochissimi carboidrati.

### Alimenti Consigliati

* **Grassi Buoni (Essenziali):** Olio extravergine d'oliva (usalo generosamente), burro, avocado, olive.
* **Frutta a guscio (con moderazione):** Noci, mandorle, noci pecan <span style='color:magenta'>(evita gli anacardi perché sono ricchi di zuccheri)</span>.
* **Formaggi stagionati:** Parmigiano Reggiano o Grana Padano (hanno zero carboidrati).
* **Bevande:** Acqua (almeno 2 litri), caffè e tè non zuccherati.

### Alimenti Fortemente Sconsigliati (I "Fuori Gioco")

* **Zuccheri:** Dolce, zucchero bianco/di canna, miele, bevande zuccherate e succhi di frutta.
* **Cereali e Amidi:** Pane, pasta, riso, pizza, mais, farro, couscous.
* **Legumi:** Fagioli, lenticchie, ceci, piselli (troppi carboidrati per la cheto).
* **Frutta:** La maggior parte va evitata. Sono ammessi solo piccoli frutti rossi (mirtilli, lamponi) in quantità minime.
* **Tuberi:** Patate e carote cotte.

## 💡 Consigli per una "Cheto Perfetta"

### 1. Non temere i grassi

Poiché stai riducendo drasticamente i carboidrati, il tuo corpo ha bisogno di un'altra fonte di energia: i **grassi**. Se mangi solo carne magra e verdure scondite, ti sentirai stanco. Condisci sempre i tuoi 100/200g di verdura con un cucchiaio di olio EVO.

### 2. Attenzione ai condimenti nascosti

<span style="color:magenta">Molte salse pronte (ketchup, aceto balsamico industriale, salse allo yogurt) contengono zucchero</span>. Usa aceto di vino o di mele, succo di limone, erbe aromatiche e spezie.

### 3. L'importanza dei Sali Minerali

Nei primi giorni potresti avvertire il "flu cheto" (un po' di mal di testa o stanchezza). È normale: il corpo sta espellendo liquidi. Assicurati di:

* Salare adeguatamente le pietanze.
* Bere molta acqua.
* <span style='color:green'>Consumare regolarmente gli **spinaci** e la **bieta** (ricchi di magnesio e potassio)</span>.

## 📅 Piano Alimentare Settimanale

### 🟢 Note fisse giornaliere:

* **Condimento:** 1 cucchiaio e mezzo (15g) di Olio EVO a pranzo e a cena. Puoi sostituire con 1/4 di Avocado (60g).
* **Integrazione:** Aggiungi 10g di semi (Chia, Lino o Sesamo) nelle insalate o come panatura.
* **Bevande:** Caffè, tè o tisane sempre senza zucchero.
